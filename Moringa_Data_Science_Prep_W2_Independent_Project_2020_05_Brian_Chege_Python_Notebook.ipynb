```
#importing libaries to be used in the analysis
import pandas as pd
import numpy as np
```

```
#Loading the datasets to be used in our analysis and converting them to dataframes
df=pd.read_csv("Telcom_dataset.csv",delimiter=',')
df1=pd.read_csv("Telcom_dataset2.csv",delimiter=',')
df2=pd.read_csv("Telcom_dataset3.csv",delimiter=',')
#Cleaning the dataset by putting all the column names to be same
df.rename(columns={"PRODUTC":'PRODUCT','DATETIME':'DATE_TIME'},inplace=True)
df2.rename(columns={"CELLID":'CELL_ID',"SIET_ID":'SITE_ID'},inplace=True)
df1.rename(columns={'DW_A_NUMBER':'DW_A_NUMBER_INT','DW_B_NUMBER':'DW_NUMBER_INT'},inplace=True)
```

```
#Concatenating the three dataframes to be one for easier analysis
dfs=[df,df1,df2]
df_s=pd.concat(dfs)
df_s
```

```
#Finding out whether there are missing values in our dataset
df_s.isnull()
```

```
#The code below picks the column of interest in the dataframe
df_interest=df_s[["PRODUCT","VALUE","DATE_TIME","CELL_ON_SITE","CELL_ID","SITE_ID"]]
df_interest
```

```
#The code below loads the cell geo dataset into our notebook and converts it into a dataframe
cf=pd.read_csv("cells_geo.csv",delimiter=',')
cf
```

```
#The code below splits the columns using the ";" delimiter
cf_new=cf[';VILLES;STATUS;LOCALISATION;DECOUPZONE;ZONENAME;LONGITUDE;LATITUDE;REGION;AREA;CELL_ID;SITE_CODE'].apply(lambda x:pd.Series([i for i in reversed(x.lower().split(';'))]))
```

```
#The code below renames the cells cells geo 
cf_new.rename(columns={3:"REGION",4:"LONGITUDE",5:"LATITUDE",6:"ZONENAME",2:"LOCALISATION",0:'SITE_ID',1:"CELL_ID",10:"CITY"},inplace=True)
```

```
#The code below picks the top few rows of the cells geo dataframe to display the changes made
cf_new.head()
```

```
#The code below picks the columns of interest in the cells_geo dataset
cf_interest=cf_new[["SITE_ID","CELL_ID","LOCALISATION","REGION","LONGITUDE","LATITUDE","CITY"]]
```

```
#The line of code below displays the columns of interest
cf_interest
```

```
#The line of code below merges the Telcom dataframe with the cells geo dataframe using the site-id
comb=df_interest.merge(cf_interest,how='left',on=["SITE_ID"])
comb
```

```
#The line of code below cleans the merged dataset by dropping the duplicate rows
ccomb=comb.drop_duplicates(["SITE_ID"],keep='last')
ccomb
```

```
#The line of code below drops the rows with less than 7 observations and contains alot of missing values
comb_clean=ccomb.dropna(thresh=7)
comb_clean
```

```
#The line of code below picks the columns of interest in the merged dataframe to do the analysis with
d_f=comb_clean[["PRODUCT","VALUE","DATE_TIME","CELL_ON_SITE","LOCALISATION","REGION","CITY"]]
d_f
```

```
#The line of code below does the mean of the product usage
d_f.groupby(['PRODUCT']).mean()
```

```
# The line of code below does the mean based on the localisations
d_f.groupby(['LOCALISATION']).mean()
```

```
#The line of code below does the basic staistics to find the mean value of usage
d_f['VALUE'].describe()
```

```
#The line of code below does a basic statistics of the value divided by localisation
groupby_product=d_f["VALUE"].groupby(d_f['LOCALISATION'])
```

```
groupby_product.describe()
```

```
#The line of code below finds the usage of each product by localisation
d_f["VALUE"].groupby([d_f["LOCALISATION"],d_f["PRODUCT"]]).describe()
```

```
#The line of code below displays the table describing the means of usage of products in tables across the different localisations
d_f.pivot_table(index=["PRODUCT","LOCALISATION"],aggfunc="count")
```

```
#The line of code below finds out the frequency of cell on site usage in the different regions
d_f["LOCALISATION"].groupby(d_f["CELL_ON_SITE"]).describe()
```

```
city=d_f["VALUE"].groupby(d_f["CITY"]).describe()
city
```
